```python
from transformers import pipeline
classifier = pipeline("sentiment-analysis")
```

```python
#Here we are using the default sentiment analysis model i.e. DistilBert uncased English model
classifier("Its very hard to start Deep Learning")
```

```python
results = classifier(["Deep learning is very intresting" , "Though deep leaning is confusing to understand it is intresting"])

for result in results:
  print(f"label:{result['label']}  , with score : {result['score']}")
```
